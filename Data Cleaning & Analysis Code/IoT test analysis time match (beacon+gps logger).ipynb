```python
import urllib.request, json
import requests
import pandas as pd
```

```python
url = "http://143.198.161.82:8086/get"
r = requests.get(url)
```

```python
dictr = r.json()
```

```python
dictr
```

```python
main = pd.DataFrame.from_dict(dictr)
main.head()
```

```python
len(main)
```

```python
beacons = main["beacons"]
from itertools import chain
beacondetails = pd.DataFrame(list(chain.from_iterable(beacons)))
beacondetails.head()
```

```python
beacondata = beacondetails["ibeacon_data"]
beacondatadf = pd.DataFrame(beacondata)
beacondatadf
```

```python
beacondatadetails = beacondatadf["ibeacon_data"].apply(pd.Series)
beacondatadetails
```

```python
detailsdf = pd.concat([main, beacondetails, beacondatadetails],axis=1)
detailsdf.head()
```

```python
cleandf = detailsdf[['timestamp','name','beacon_type','distance','hashcode','isBlocked','last_seen','manufacturer','rssi','tx_power','eddystoneUidData','eddystone_url_data','telemetry_data','major','minor','uuid']]
cleandf.head()
```

```python
len(cleandf)
```

```python
finaldf = cleandf[cleandf["timestamp"].notnull()]
finaldf.head()
```
